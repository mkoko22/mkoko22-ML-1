```python
import pandas as pd
run_id = "c89bcf9392584f3e93c0adc3dfa248be"
artifact_path = "test_df_with_ids.csv"
local_path = mlflow.artifacts.download_artifacts(run_id=run_id, artifact_path=artifact_path)

test_df_scaled = pd.read_csv(local_path)
print("Loaded Scaled Test Data:")
print(test_df_scaled.shape)
'Id' in test_df_scaled.columns
```

```python
import mlflow
import mlflow.sklearn
import pandas as pd

# Load the model
model_name = "HousePricePrediction_BestModel"
model_version = 2

model_uri = f"models:/{model_name}/{model_version}"

loaded_model = mlflow.sklearn.load_model(model_uri)

feature_importances = loaded_model.feature_importances_

test_columns = test_df_scaled.columns.tolist()

feature_importance_dict = dict(zip(test_columns, feature_importances))
sorted_features = sorted(feature_importance_dict.items(), key=lambda x: x[1], reverse=True)

top_10_features = [feature[0] for feature in sorted_features[:10]]

test_df_scaled_selected = test_df_scaled[top_10_features]

test_preds = loaded_model.predict(test_df_scaled_selected)

submission_df = pd.DataFrame({
    'Id': test_df_scaled['Id'],
    'SalePrice': test_preds
})

submission_df.to_csv('submission.csv', index=False)

print("Model loaded and predictions saved to 'submission.csv'")
```
